```
# Mthod to get Image
def getImage(path,index):
  # Get one test image for the 
  path = os.path.join(path, all_imagesNames[index])
  img = cv2.imread(path)
  RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  rescaled = cv2.resize(RGB_img, (img_sizeX, img_sizeY))
  rescaled = rescaled.astype('float32') / 255
  return rescaled

def getNpy(path, index):

    from numpy import loadtxt
    # load array
    name = all_imagesNames[index]
    imageName = name.split('.')[0]
    path = os.path.join(path, imageName )
    data = np.load(path + ".npy")
    return data
```

```
# Set variables
# initialize the number of epochs and batch size
EPOCHS = 50
BS = 28

trainImageCount = 4500
validationCount = len(all_imagesNames) -4500

# Create generator
def image_generator(bs, startPos, endPos,mode="train" ):
    # Set the index
    index = startPos

    while True:
        # Initialize our image batches.
        startImages = []
        gabor_one_Images = []
        gabor_two_Images = []
        gabor_three_Images = []

        # keep looping until we reach our batch size
        while (len(startImages) < bs):
            if(index == endPos):
                index = startPos

            startImages.append(np.array(getImage(main_path, index)))
            gabor_one_Images.append(getNpy(gabor_one_path, index))
            gabor_two_Images.append(getNpy(gabor_two_path, index))
            gabor_three_Images.append(getNpy(gabor_three_path, index))

            index +=1


        startImages = np.reshape(startImages, (len(startImages), img_sizeY, img_sizeX, 3))
        gabor_one_Images = np.reshape(gabor_one_Images, (len(gabor_one_Images), img_sizeY, img_sizeX, 1))
        gabor_two_Images = np.reshape(gabor_two_Images, (len(gabor_two_Images), img_sizeY, img_sizeX, 1))
        gabor_three_Images = np.reshape(gabor_three_Images, (len(gabor_three_Images), img_sizeY, img_sizeX, 1))



        yield startImages, { "decoded" : startImages, "gabor1" : gabor_one_Images, "gabor2" : gabor_two_Images, "gabor3" :gabor_three_Images}
```

```
# Create the generators
trainGen = image_generator(BS,0,4500)
testGen = image_generator(BS, 4500,len(all_imagesNames))
```
